```python
from ultralytics import YOLO
#load a model
model = YOLO ("yolov8n.yaml") #build a new model from scratch
model = YOLO ("yolov8n.pt") #load a pretrained model (recommended for traning)

#use the model
results = model.train(data= "coco128.yaml", epochs=3) #train the model
results = model.val() #evaluated model performance on the validation set
results = model("https://ultralytics.com/images/bus.jpg") # predict on the image
success = model.export(format="onnx") # export the model on ONNX format
```
